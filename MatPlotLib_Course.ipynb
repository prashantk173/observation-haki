```python
x = np.linspace(0,5,11) #linspace is used for linearly placed values
y = x**2
```

Functional Method

#*Using Object Oriented Method*

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Titles')

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.7])
axes1.plot(x,y)
axes2 = fig.add_axes([0.2,0.4,0.4,0.3])
axes2.plot(y,x)
axes1.set_title('Big')
axes1.set_xlabel('xside')
axes1.set_ylabel('yside')
axes2.set_title('small')
axes2.set_xlabel('xline')
axes2.set_ylabel('yline')

plt.show()
```

#*Creating Subplots using Object oriented method*

```python
fig,axes = plt.subplots(nrows=1,ncols=2) # we can mention the number of subplots needed using nrows and ncols method.
```

```python
figax = plt.subplots(nrows=3,ncols=4)
# to reduce the overlapping we use tight fn.
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
  current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#we can index the axes object such as
axes[0].plot(x,y)
axes[0].set_title('First_plot')
axes[1].plot(y,x)
axes[1].set_title('Second_plot')
plt.show()
plt.tight_layout()

#also we can titles and labels
```

```python
axes # this is a array of the axes also called tuple unpacking
```

#*Figure Size and DPI*

```python
fig = plt.figure(figsize=(3,5), dpi=80)

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('Hello World')
plt.show()
```

```python
# the same can be done with subplots
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(6,2), dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

#giving Legend to the plot

```python
fig = plt.figure(figsize=(8,5))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.plot(y,x)
axes.set_title('for legend')
axes.set_xlabel('xline')
axes.set_ylabel('yline')
plt.show()
```

```python
 fig = plt.figure(figsize=(8,5))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,'r', label= 'X Squared')
axes.plot(x,x**3,'g', label= 'X Cubed')
axes.set_title('for legend')
axes.set_xlabel('xline')
axes.set_ylabel('yline')
axes.legend(loc=0)
plt.show()
# Here we have given the axes therir labels which will be reconized by the legend fn. and then it will return
#and also for the best place of the legend to be located put that in loc=0, and also the position can be changed using x,y value in a tuple - (x,y)
```

```python
fig.savefig('for Legend.png', dpi = 200)
```

#*Customizing the data*

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green', lw = 3, alpha=0.8,ls='--',marker='o',markersize=20, 
        markerfacecolor='red',markeredgewidth=3,markeredgecolor='blue') #RGB hex codes search in google for more colors.
plt.show()
```

```python
 fig = plt.figure(figsize=(6,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'green', lw = 3, alpha=0.8,ls='--')
ax.set_xlim([0,1]) #Setting the xlimit
ax.set_ylim([0,2]) #Setting the ylimit

plt.show()
```
