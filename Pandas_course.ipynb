#**Groupby**

```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
# creating a dataframe sample
data = {'Company':['GOOG','GOOG','APPLE','APPLE','MSFT','MSFT'], 
        'Name':['sam','charlie','chidi','rose','faith','TENNYSON'],
        'Sales':[100,120,350,30,240,400]}
```

```python
datadf = pd.DataFrame(data)
```

```python
datadf
```

```python
bycomp = datadf.groupby('Company') #grouping the data based on compnay as it is having the common items.
```

```python
bycomp # this gives the location of the object grouped.
```

```python
# we can use this and put methods on it
bycomp.mean()
```

```python
#similarly the standard deviation and sum
bycomp.sum()
```

```python
bycomp.std()
```

```python
#as we can see these are natural datframeswe can use loc and call function and use dataframe opsand methods on them
bycomp.sum().loc['MSFT']
```

```python
datadf.groupby('Company').sum().loc['MSFT'] #all the same things in just one line.
```

```python
 # we can also use functions such as count, min, max amd other 
 #arithmetic function on the numericals and make note that these operations are not valid on strings.
 datadf.groupby('Company').count()
```

```python
datadf.groupby('Company').min()
```

```python
datadf.groupby('Company').sum().max()
```

```python
datadf
```

```python
 # we can use the groupby with another method called describe method which is commonly used and gives all the aggregate data.
 datadf.groupby('Company').describe()
```

```python
 datadf.groupby('Company').describe().transpose() #transpose can be used to change the format.
```
